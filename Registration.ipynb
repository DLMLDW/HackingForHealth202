Notebook to register two images 

```python
import cv2 
import numpy as np 
from skimage.io import imread, imsave
import matplotlib.pyplot as plt
from pystackreg import StackReg 
```

```python
path = "./Segmentation/"
image1 = "00028999_binary.tiff"
image2 = "00015999_binary.tiff"
```

```python
img1 = imread(path + image1).astype("float")
img2 = imread(path + image2).astype("float")


print("Imported two images of shape: ", np.shape(img1), " and ", np.shape(img2))
```

```python
sr = StackReg(StackReg.BILINEAR)
sr.register(img1, img2)
out = sr.transform(img2)
```

```python
plt.imshow(img1)
plt.title("Groundtruth Image")
plt.show()
plt.imshow(img2)
plt.title("Image to Transform")
plt.show()
plt.imshow(out)
plt.title("Transformed Image")
plt.show()
```
